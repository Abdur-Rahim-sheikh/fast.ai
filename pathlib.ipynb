```python
from pathlib import Path
```

```python
file = Path("myfile.txt")
folder = Path("temp")
file.touch()
folder.mkdir()
```

```python
# touch, mkdir
name = folder / "myfile.txt"
name.touch()
```

```python
for path in Path(".").iterdir():
    print(path.name)
```

```python
# recursive
for path in Path(".").glob("**/*"):
    print(path)
```

```python
# deleting
for path in Path(".").iterdir():
    if path.name.startswith("my"):
        path.unlink()
```
